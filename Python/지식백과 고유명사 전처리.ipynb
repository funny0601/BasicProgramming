```python
noun_dic=data1['단어'].tolist()
```

```python
def splitToNNP(text):
    num=[]
    NNP_new_list=[]
    del_text_list=[]
    
    for k in range(0, len(text)):
        pos_tagger=komoran.pos(text[k]) 
        #print(pos_tagger)

        NNP_list=[word[0] for word in pos_tagger if word[1]=='NNP' and len(word[0])>1 ]##고유명사이고 단어길이 2 이상인 것들만 추출.. 
        length=len(NNP_list) ##NNP 개수 저장
        #print(length)
        print("1차 -고유명사만 뽑음:", NNP_list)
        
        del_list=[]
        for i in range(0, length):
            count=1 #몇 번 중복된건지 세는 거.. 우선 한 번 이상은 나오니까 무조건 1로 초기화
            for j in range(i+1, length):
                if NNP_list[i] == NNP_list[j]:
                    count=count+1
            if(count>=2):  #단어가 2번 이상 반복되었다면 그 인덱스를 리스트에 저장해놓기
                del_list.append(str(i))
        del_list.reverse()  #중복되는 index뽑아서 저장하고 역으로 저장 -> 그래야 삭제해도 리스트 인덱스 변하지 않음 
        
        for i in range(0, len(del_list)):
            del_index=int(del_list[i])
            #print(NNP_list[del_index], del_index)
            del(NNP_list[del_index]) #삭제
        print("2차 -중복명사를 제거:", NNP_list)
        
        for i in range(0, len(NNP_list)): # 고유명사 사전과 동일할 경우 지운다
            for j in range(0, len(noun_dic)):
                if NNP_list[i] == noun_dic[j]:
                    NNP_list[i]=0 #우선 0으로 처리
                    break
                
        NNP_list = [e for e in NNP_list if e!=0] #0인것만 싹다 지움
        print("3차 -보통명사를 제거:",NNP_list)
                 
        # 고유명사 사전과 동일할 경우 지운다
        if len(NNP_list)==0:
            NNP_list.append('none')
        
        print("4차 -명사0개를 명시:",NNP_list)

        length=len(NNP_list) ##중복된 것 삭제되었으니까 길이 재정의
        #print(length)
        
       
        if (length >=1 ): # 이거 해 줄 필요 없었는데.. 그냥 지우기 귀찮으니깐
            rand_num=random.randint(0, length-1)
        #print(rand_num)
            NNP_new_list.append(NNP_list[rand_num])
        #print(k)
            
            entity_name="@"+str(k+1)
        #print(entity_name)
            num.append(entity_name)
        #print(k)
            text[k]=text[k].replace(NNP_list[rand_num],num[k]) ##고유명사를 entity로 바꾸기
            
    return text, NNP_new_list
```

```python
result, entity_list=splitToNNP(text)
#print(entity_list)
```
